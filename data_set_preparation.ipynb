### Features Implemented
1. Reads all csv files in a folder 
2. Can show basic statistics for all of them

### Importing libraries

```python
import numpy as np
import pandas as pd
import os
```

### Reading files

```python
folder_path = '/Users/antique/Downloads/bihs_usaid_4.23.2013/household/data/csv/'
file_name_list=sorted(os.listdir(folder_path))
num_of_files =len(file_name_list);
```

```python
#keeping file name in array without extension
for index in range(num_of_files):
    only_name = os.path.splitext(file_name_list[index])[0]
    file_name_list[index] = only_name
print(file_name_list[0:5]) #printing first five file names
```

### Reading all files as Panda_Data_Frame and Keeping reference in array

```python
file_list=[]
for index in range(num_of_files):
    file_reading_path = folder_path+file_name_list[index]+'.csv'
    file_list.append(pd.read_csv(file_reading_path))
```

### Printing File Name, First Row of each file and Basic Stat

```python
for index in range(num_of_files):
    print('\nFile name:',file_name_list[index],)
    print('\nFirst row: \n',file_list[index].iloc[:1])
    print('\nStatistics: \n',file_list[index].describe())
    
print() #this cell prints nothing
```
